```python
unique_trainx = train.user_id.unique()  
trainx = {o:i for i, o in enumerate(unique_trainx)}

# Get int mapping for isbn in train dataset
unique_trainy = train.isbn.unique()  
trainy = {o:i for i, o in enumerate(unique_trainy)}
```

```python
# Get int mapping for user_id in test dataset
unique_testx = test.user_id.unique()  
testx = {o:i for i, o in enumerate(unique_testx)}

# Get int mapping for isbn in test dataset
unique_testy = test.isbn.unique() 
testy = {o:i for i, o in enumerate(unique_testy)}
```

```python
# training set
train['u_unique'] = train['user_id'].map(trainx)
train['i_unique'] = train['isbn'].map(trainy)

# testing set
test['u_unique'] = test['user_id'].map(testx)
test['i_unique'] = test['isbn'].map(trainy)

# Convert back to three feature of dataframe 
train = train[['u_unique', 'i_unique', 'book_rating']]
test= test[['u_unique', 'i_unique', 'book_rating']]
```
