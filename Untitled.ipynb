```python
def popular_books(train,n=100):
    rating_count=train.groupby("Book-Title").count()["Book-Rating"].reset_index()
    rating_count.rename(columns={"Book-Rating":"NumberOfVotes"},inplace=True)
    
    rating_average=train.groupby("Book-Title")["Book-Rating"].mean().reset_index()
    rating_average.rename(columns={"Book-Rating":"AverageRatings"},inplace=True)
    
    popularBooks=rating_count.merge(rating_average,on="Book-Title")
    
    def weighted_rate(x):
        v=x["NumberOfVotes"]
        R=x["AverageRatings"]
        
        return ((v*R) + (m*C)) / (v+m)
    
    C=popularBooks["AverageRatings"].mean()
    m=popularBooks["NumberOfVotes"].quantile(0.90)
    
    popularBooks=popularBooks[popularBooks["NumberOfVotes"] >=250]
    popularBooks["Popularity"]=popularBooks.apply(weighted_rate,axis=1)
    popularBooks=popularBooks.sort_values(by="Popularity",ascending=False)
    return popularBooks[["Book-Title","NumberOfVotes","AverageRatings","Popularity"]].reset_index(drop=True).head(n)
```

```python
n = 10
top_ten = pd.DataFrame(popular_books(train, n))

for i in range(len(top_ten)):
    title = top_ten.iloc[i]['Book-Title']
    author = train[train["Book-Title"]==title]["Book-Author"].iloc[0]
    rating = round(train[train["Book-Title"]==title]["Book-Rating"].mean(), 1)
    print(f'{i+1}. Title: {title} | Author: {author} | Rating: {rating}')
```

```python
from sklearn.metrics import mean_absolute_error
import numpy as np

def calculate_mae(predicted_titles, predicted_ratings):
    # Initialize empty lists to store the actual ratings and corresponding predicted ratings
    actual_ratings = []
    matched_predicted_ratings = []
    
    # Iterate over the predicted book titles and ratings
    for title, rating in zip(predicted_titles, predicted_ratings):
        # Find the corresponding row in the test data based on the predicted title
        test_row = test[test['Book-Title'] == title]
        
        # Check if the row is found in the test data
        if not test_row.empty:
            # Retrieve the actual rating from the test data
            actual_rating = test_row['Book-Rating'].values[0]
            
            # Append the actual rating and corresponding predicted rating to their respective lists
            actual_ratings.append(actual_rating)
            matched_predicted_ratings.append(rating)
    
    # Convert the actual ratings and matched predicted ratings lists to NumPy arrays
    actual_ratings = np.array(actual_ratings)
    matched_predicted_ratings = np.array(matched_predicted_ratings)
    
    # Calculate MAE
    mae = mean_absolute_error(actual_ratings, matched_predicted_ratings)
    
    # Return the MAE
    return mae

# Assuming you have the predicted book titles, predicted ratings, and test data DataFrame 'test' with columns 'Book-Title' and 'Book-Rating'
predicted_titles = ['The Secret Life of Bees', 'The Da Vinci Code', 'The Lovely Bones A Novel',
                    'The Red Tent Bestselling Backlist', 'Divine Secrets of the Ya Ya Sisterhood A Novel',
                    'Bridget Jones\'s Diary', 'A Painted House', 'The Nanny Diaries A Novel', 'Wild Animus']
predicted_ratings = [8.5, 8.4, 8.2, 8.2, 7.8, 7.6, 7.5, 7.4, 4.4]

# Call the function to calculate MAE
mae = calculate_mae(predicted_titles, predicted_ratings)

# Print the MAE
print("MAE:", mae)
rmse = mae*np.sqrt(len(predicted_titles))
print("rmse:", rmse)


```

```python
from scipy.stats import pearsonr

def calculate_similarity(matrix):
    n_books = matrix.shape[0]
    similarities = np.zeros((n_books, n_books))
    
    for i in range(n_books):
        for j in range(i, n_books):
            similarity, _ = pearsonr(matrix[i], matrix[j])
            similarities[i, j] = similarity
            similarities[j, i] = similarity
    
    return similarities

pearsonsimilarities = calculate_similarity(books_table.values)
```

```python
def recommendation(book):
    index = np.where(books_table.index == book)[0][0]
    item_indices = np.argsort(pearsonsimilarities[index])[::-1][1:6]
    data = []

    for item_index in item_indices:
        item = []
        recommended_book = books_table.index[item_index]
        temp_table = books[books['Book-Title'] == recommended_book]
        item.extend(list(temp_table.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_table.drop_duplicates('Book-Title')['Book-Author'].values))
      
        data.append(item)

    return data

```
